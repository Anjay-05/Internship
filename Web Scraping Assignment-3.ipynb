```python
#Creating list to store data
Name=[]
Distance=[]
Rating=[]
TReview=[]
OReview=[]
Pprice=[]
Dprice=[]
Facilities=[]
```

```python
#Scraping the Hostel names
try:
    name=driver.find_elements_by_xpath('//div[@class="property"]/div[1]/h2')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('--')
    
#Scraping the Distance from city centre
try:
    distance=driver.find_elements_by_xpath('//div[@class="property"]/div[1]/div[1]/a/span[1]')
    for i in distance:
        Distance.append(i.text.split('-')[1])
except NoSuchElementException:
    Distance.append('--')
    
#Scraping the Ratings
try:
    rating=driver.find_elements_by_xpath('//div[@class="property"]/div[2]/a/div[1]/div[1]')
    for i in rating:
        Rating.append(i.text)
except NoSuchElementException:
    Rating.append('--')
    
#Scraping the Total Review
try:
    treview=driver.find_elements_by_xpath('//div[@class="property"]/div[2]/a/div[1]/div[2]/div[2]')
    for i in treview:
        TReview.append(i.text)
except NoSuchElementException:
    TReview.append('--')
    
#Scraping the Overall Review
try:
    oreview=driver.find_elements_by_xpath('//div[@class="property"]/div[2]/a/div[1]/div[2]/div[1]')
    for i in oreview:
        OReview.append(i.text)
except NoSuchElementException:
    OReview.append('--')
    
#Scraping the Privates from price
try:
    price=driver.find_elements_by_xpath('//div[@class="prices-col"]/a/div[1]')
    for i in price:
        Pprice.append(i.text.replace('\n',' '))
except NoSuchElementException:
    Pprice.append('--')
    
#Scraping the Dorms from price
try:
    dprice=driver.find_elements_by_xpath('//div[@class="prices-col"]/a/div[2]')
    for i in dprice:
        Dprice.append(i.text.replace('\n',' '))
except NoSuchElementException:
    Dprice.append('--')
    
#Storing urls of Hostels
URL=[]
urls=driver.find_elements_by_xpath('//h2[@class="title title-6"]/a')
for i in urls:
    URL.append(i.get_attribute('href'))
for j in URL:
    driver.get(j)
    
    #Scraping the Facilities
    try:                                                                      
        faci=driver.find_elements_by_xpath('//div[@class="facilities-container"]/ul')
        for k in faci:
            Facilities.append(k.text.replace('\n',' '))

    except NoSuchElementException:
        Facilities.append('--')   
```

```python
len(Facilities),len(Name),len(Distance),len(Rating),len(TReview),len(OReview),len(Pprice),len(Dprice)
```

```python
#Making Dataframe
df=pd.DataFrame({'Hostel Name':Name,
                'Hostel Distacne from city centre':Distance,
                'Rating':Rating,
                'Total Review':TReview,
                'Overall Review':OReview,
                'Private from price':Pprice,
                'Dorms from price':Dprice,
                'Facilities':Facilities})
df
```
